# Cubic Regression with Gradient Descent

### Outline of Program
1. Get cubic-shaped noisy dataset
2. Generate random starting cubic function
3. Evaluate error of starting cubic function using Mean Squared Error
4. Use Gradient Descent to reduce error

```python
import random
import numpy as np
import matplotlib.pyplot as plt # For graphs
```

```python
# Variables to change
xs_size = 1000 # Size of inputs
noise = 100   # Amount of noise
lr = 0.000001 # Learning rate (step of gradient descent)
```

```python
def evaluate(coeffs, x):
    """
    Evaluate y-value using given coefficients and input
    
    args:
        coeffs: List of coefficients for polynomial
        x: Input value to be evaluated
        
    """
    
    return coeffs[0]*x**3 + coeffs[1]*x**2 + coeffs[2]*x + coeffs[3]

def noise_function(coeffs, x, noise):
    """
    Return the y-value above with noise added
    
    args:
        coeffs: List of coefficients for polynomial
        x: Input value to be evaluated
        noise: Size of noise
        
    """
    
    return evaluate(coeffs,x) + np.random.normal() * noise
```

```python
coeffs = [1,2,3,4]

xs = np.random.uniform(-10,10,xs_size)
ys = noise_function(coeffs, xs, noise)
    
plt.figure(figsize=(20,10))
plt.plot(xs,ys,'g+')
plt.title('Original Dataset')
plt.show()
```

```python
rand_coeffs = [random.randrange(-10,10),random.randrange(-10,10),random.randrange(-10,10),random.randrange(-10,10)]
```

```python
predicted_ys = evaluate(rand_coeffs,xs)
    
plt.figure(figsize=(20,10))
plt.plot(xs,predicted_ys,'r+', label="Predicted Values")
plt.plot(xs,ys,'g+', label="Original Values")
plt.title("Original Dataset vs Starting Predicted Values")
plt.legend(loc="upper right")
plt.show()
```

```python
def mse(original_ys, predicted_ys):
    """
    Measure the error of predicted y values
    
    args:
        original_ys: List containing original dataset values
        predicted_ys: List containing predicted y-values
        
    return:
        error: The mean squared error of our predicted values
        
    """
    
    return sum( (original_ys - predicted_ys) **2 ) / len(original_ys)

initial_model_loss = mse(ys,predicted_ys)
print(f"Initial Mean Squared Error: {initial_model_loss}")
```

```python
def batch_gradient_descent(coeffs, xs, original_ys, predicted_ys):
    
    a_s = []
    b_s = []
    c_s = []
    d_s = []
    
    xs_ys_ybars = list(zip(xs,original_ys,predicted_ys)) # Put xs,ys,ybars into a list of 3-tuple
    n = len(xs)
    
    # Get the gradient of coefficients
    for x,y,ybar in xs_ys_ybars:
        a_s.append( x**3 * (y - ybar) )
        b_s.append( x**2 * (y - ybar) )
        c_s.append( x * (y-ybar) )
        d_s.append(y-ybar)
        
    gradient_a = (-2/n) * sum(a_s) # Gradient of coefficient a
    gradient_b = (-2/n) * sum(b_s) # Gradient of coefficient b
    gradient_c = (-2/n) * sum(c_s) # Gradient of coefficient c
    gradient_d = (-2/n) * sum(d_s) # Gradient of coefficient d
    
    # New coefficients
    a = coeffs[0] - gradient_a * lr # New coefficient a
    b = coeffs[1] - gradient_b * lr # New coefficient b
    c = coeffs[2] - gradient_c * lr # New coefficient c
    d = coeffs[3] - gradient_d * lr # New coefficient d
    
    new_ybars = evaluate([a,b,c,d], xs) # New predicted y values
    
    new_mse = mse(original_ys,new_ybars) # New mean squared error
    
    return [a,b,c,d], new_ybars, new_mse 
```

```python
def gd_iterations(iterations,coeffs, xs, original_ys, predicted_ys):
    new_coeffs = []
    new_ybars = []
    new_mse = 0
    
    for _ in range(iterations):
        new_coeffs, new_ybars, new_mse = batch_gradient_descent(coeffs,xs,original_ys,predicted_ys)
        
    return new_coeffs, new_ybars, new_mse
```

```python
coeffs, ybars, mse = gd_iterations(1000, rand_coeffs, xs, ys, predicted_ys)

plt.figure(figsize=(20,10))
plt.plot(xs,ys,'g+', label='Original Dataset')
plt.plot(xs,ybars,'r+', label='Batch Gradient Descent')
plt.title('Original Data vs Gradient Descent Predicted Values')
plt.legend(loc='upper right')
plt.show()
```
